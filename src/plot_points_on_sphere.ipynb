```python
%matplotlib inline
import numpy as np
import Bio
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# pymol center of mass calculated for 1uaz_tm.pdb
com_coordinates = [0.186, -0.183, 0.268]
```

```python
num_pts = 250
# Generate n indices on an evenly distributed interval
indices = np.arange(0, num_pts, dtype=float) + 0.5
golden_angle = np.pi * (3 - np.sqrt(5))
# We need only points on an hemisphere instead of whole sphere.
phi = np.arccos(1 - indices / num_pts)
theta = golden_angle * indices

# Translate the spherical coordinates to cartesian 3D (x, y, z)
# Center the hemisphere to the center of mass of the protein
points = np.zeros((num_pts, 3))
points[:, 0] = com_coordinates[0] + np.cos(theta) * np.sin(phi)
points[:, 1] = com_coordinates[1] + np.sin(theta) * np.sin(phi)
points[:, 2] = com_coordinates[2] + np.cos(phi)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2])
plt.show()
```
