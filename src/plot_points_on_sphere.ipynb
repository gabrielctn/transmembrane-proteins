```python
import numpy
```

```python
# pymol center of mass calculated for 1uaz_tm.pdb
com_coordinates = [0.186, -0.183, 0.268]
```

```python
n = 256
golden_angle = numpy.pi * (3 - numpy.sqrt(5))
theta = golden_angle * numpy.arange(n)
z = numpy.linspace(1 - 1.0 / n, 1.0 / n - 1, n)
radius = numpy.sqrt(1 - z * z)

com_x, com_y, com_z = com_coordinates
points = numpy.full((n, 3), [com_x, com_y, com_z])
points[:,0] = radius * numpy.cos(theta)
points[:,1] = radius * numpy.sin(theta)
points[:,2] = z
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2])
plt.show()
```
